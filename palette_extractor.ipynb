```python
def RGB_format(colour_tuple):
    return ('%02x%02x%02x' % colour_tuple).upper()
```

```python
RGB_format(tuple((255, 255, 255)))
```

```python
RGB_palette = [RGB_format(tuple(int(i) for i in j)) for j in df.values]
```

```python
def RGB_palette(number):
    df = pandas.read_csv(f"../palettes/Craig/data/palette_{number}.csv", header=None)
    palette = [RGB_format(tuple(int(i) for i in j)) for j in df.values]
    final_string = ""
    for i, colour in enumerate(palette):
        final_string += colour
        if i % 32 == 31:
            final_string += '\n'
    return final_string
```

```python
print(RGB_palette(13))
```
